
1. What is statistics, and why is it important?  
- Statistics is a way of collecting, organizing, and understanding data. It's important because it helps us make better decisions using facts instead of guesses.

2. What are the two main types of statistics?  
- Descriptive statistics (describe data)  
- Inferential statistics (make predictions or conclusions)

3. What are descriptive statistics?  
- They summarize dataâ€”for example, showing the average or range of values.

4. What is inferential statistics?  
- It helps us make guesses or conclusions about a larger group using data from a smaller group (sample).

5. What is sampling in statistics?  
- Sampling means picking a small group from a big group so we can study it and learn about the bigger one.

6. What are the different types of sampling methods?  
- Random sampling  
- Systematic sampling  
- Stratified sampling  
- Cluster sampling

7. What is the difference between random and non-random sampling?  
- Random: Everyone has an equal chance of being picked.  
- Non-random: Selection is not random, so it might be biased.

8. Define and give examples of qualitative and quantitative data.  
- Qualitative: Describes qualities (e.g., color = red, blue)  
- Quantitative: Involves numbers (e.g., age = 25)

9. What are the different types of data in statistics?  
- Nominal  
- Ordinal  
- Interval  
- Ratio

10. Explain nominal, ordinal, interval, and ratio levels of measurement.  
- Nominal: Categories (e.g., gender)  
- Ordinal: Order matters (e.g., ranks)  
- Interval: Equal spacing, no true zero (e.g., temperature)  
- Ratio: Has a true zero (e.g., weight)

11. What is the measure of central tendency?  
- It shows the center or average of a dataset. The main ones are mean, median, and mode.

12. Define mean, median, and mode.  
- Mean: Average  
- Median: Middle value  
- Mode: Most frequent value

13. What is the significance of the measure of central tendency?  
- It helps understand what a "typical" value in a dataset looks like.

14. What is variance, and how is it calculated?  
- Variance shows how spread out numbers are. It's the average of the squared differences from the mean.

15. What is standard deviation, and why is it important?  
- It shows how much values typically differ from the mean. It helps understand consistency.

16. Define and explain the term range in statistics.  
- Range = Highest value - Lowest value. It shows how spread out the data is.

17. What is the difference between variance and standard deviation?  
- Variance = average of squared differences  
- Standard deviation = square root of variance

18. What is skewness in a dataset?  
Skewness tells if the data is lopsided or uneven.

19. What does it mean if a dataset is positively or negatively skewed?  
- Positive skew = tail on the right  
- Negative skew = tail on the left

20. Define and explain kurtosis.  
- Kurtosis shows if the data is more peaked or has heavy tails compared to normal data.

21. What is the purpose of covariance?  
- It shows if two variables move together. If one increases, does the other also increase?

22. What does correlation measure in statistics?  
- Correlation measures the strength and direction of the relationship between two things.

23. What is the difference between covariance and correlation?  
- Covariance: Shows direction of relationship  
- Correlation: Shows direction and strength, ranges between -1 and 1

24. What are some real-world applications of statistics?  
Used in:  
- Healthcare (disease tracking)  
- Business (sales trends)  
- Politics (election polls)  
- Sports (player stats)  
- Education (test scores)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import skew, kurtosis, zscore, pearsonr, spearmanr
import random
import statistics

# 1. Mean, Median, Mode
data = [1, 2, 3, 4, 5, 5]
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", statistics.mode(data))

# 2. Variance and Standard Deviation
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# 3. Data Types
data_types = {
    "nominal": ["Red", "Blue", "Green"],
    "ordinal": ["Small", "Medium", "Large"],
    "interval": [10, 20, 30],
    "ratio": [0, 5, 10]
}

# 4. Sampling Techniques
full_data = list(range(100))
random_sample = random.sample(full_data, 10)

df_sample = pd.DataFrame({'Age': range(18, 60), 'Group': ['A']*21 + ['B']*21})
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_sample, stratify=df_sample['Group'], test_size=0.3)

# 5. Range
print("Range:", max(data) - min(data))

# 6. Histogram
plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()

# 7. Skewness and Kurtosis
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))

# 8. Positive and Negative Skewness
data_pos = [1, 2, 3, 4, 10]
data_neg = [1, 8, 9, 10, 11]
print("Positive Skew:", skew(data_pos))
print("Negative Skew:", skew(data_neg))

# 9. Covariance
x = [1, 2, 3]
y = [4, 5, 6]
print("Covariance:", np.cov(x, y)[0][1])

# 10. Correlation Coefficient
print("Correlation Coefficient:", np.corrcoef(x, y)[0][1])

# 11. Scatter Plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()

# 12. Random vs Systematic Sampling
step = len(full_data) // 10
systematic_sample = full_data[::step]

# 13. Grouped Data Stats
grouped_data = [5, 10, 10, 15, 20]
print("Mean:", np.mean(grouped_data))
print("Median:", np.median(grouped_data))
print("Mode:", statistics.mode(grouped_data))

# 14. Simulated Data Stats
simulated_data = np.random.normal(50, 10, 100)
print("Simulated Mean:", np.mean(simulated_data))
print("Simulated Std Dev:", np.std(simulated_data))

# 15. Descriptive Stats
df = pd.DataFrame({'Marks': [70, 85, 60, 90, 95]})
print(df.describe())

# 16. Boxplot
sns.boxplot(df['Marks'])
plt.title("Boxplot")
plt.show()

# 17. Interquartile Range
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1
print("IQR:", IQR)

# 18. Z-score Normalization
df['Z_score'] = zscore(df['Marks'])
print(df)

# 19. Compare Standard Deviations
data1 = [10, 12, 14, 16, 18]
data2 = [5, 25, 35, 45, 55]
print("Std Dev of Data 1:", np.std(data1))
print("Std Dev of Data 2:", np.std(data2))

# 20. Covariance Heatmap
df_cov = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})
sns.heatmap(df_cov.cov(), annot=True, cmap='coolwarm')
plt.title("Covariance Heatmap")
plt.show()

# 21. Correlation Matrix
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Matrix")
plt.show()

# 22. Variance and Std Dev Again
data = [4, 5, 8, 9, 10]
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# 23. Skewness and Kurtosis Plot
sns.histplot(data, kde=True)
plt.title("Distribution with Skewness & Kurtosis")
plt.show()
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))

# 24. Pearson and Spearman Correlations
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)
print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
